```python
url= "https://utamavs.com/sports/womens-basketball/stats/2021-22"
import requests
import pandas as pd 
from bs4 import BeautifulSoup
data = requests.get(url)
```

```python
data.text
```

```python
html= BeautifulSoup(data.text, 'html.parser')
```

```python
html
```

```python
for script in html(["script", "style"]):   # clean the html view 
    script.extract()
```

```python
html
```

```python
html.findAll("table") # list of tables from webpage 
```

```python
 from IPython.display import display,HTML  # function to view webpage 
```

```python
display(HTML(str(html.findAll("table")[5])))[0] #View table of html 
```

```python
wbb22_average=pd.read_html(str(html.findAll("table")[5]))[0]
```
